```python
# Strings are immutable objects and cannot be modified in-place:
a = "hello"
print(a[0])
print(a[-1])

a = "hello, world!"
print(a[3:6]) # 3rd to 6th (excluded) elements: elements 3, 4, 5

print(a[2:10:2]) # Syntax: a[start:stop:step]

print(a[::3]) # every three characters, from beginning to end

a.replace('l', 'z', 1) # replace first occurrence of 'l' with 'z'
print(a)
```
